```python
print(df.market_category.unique())
# df.market_category = df.market_category.apply(lambda x: x.split(',') if pd.notna(x) else 'Nan')

df['market_category'] = df['market_category'].apply(
    lambda x: [item.strip() for item in x.split(',')] if pd.notna(x) else ['no_category'])
```

```python
market_category = set()
for i in df.market_category:
    if type(i) is not int:
        market_category.update(i)
```

```python
category_dict = {cat:idx+1 for idx, cat in enumerate(market_category)}
category_dict["Nan"] = 0

```

```python
print(category_dict)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
one_hot_encoded = mlb.fit_transform(df.market_category)
market_category_df = pd.DataFrame(one_hot_encoded, columns=mlb.classes_)

```

```python
print(market_category_df.head())
print(df.market_category.head())
```

```python
# from sklearn.preprocessing import MultiLabelBinarizer
df['engine_fuel_type'] = df['engine_fuel_type'].apply(
    lambda x: [item.strip() for item in x.split(',')] if pd.notna(x) else ['unknown']
)
mlb = MultiLabelBinarizer()
one_hot_encoded = mlb.fit_transform(df.engine_fuel_type)
fuel_type_df = pd.DataFrame(one_hot_encoded, columns=mlb.classes_)
```

```python
# maybe combine unleaded fuel type to premium, non premium
print(fuel_type_df.head())
print(df.engine_fuel_type.value_counts())
```

```python
print(sum(np.bitwise_and(market_category_df.diesel , fuel_type_df.diesel)))
print(sum(market_category_df.diesel))
print(sum(fuel_type_df.diesel))
```

```python
# combine missed diesel category through combining market category and fuel_type
fuel_type_df.diesel = np.bitwise_or(market_category_df.diesel , fuel_type_df.diesel)
```

```python
market_category_df.drop(columns='diesel')
```

```python
df = df.drop(columns= ['market_category', 'engine_fuel_type'])
combine_df = pd.concat([df,fuel_type_df], axis=1)
combine_df = pd.concat([combine_df,market_category_df], axis = 1)
```

```python
string_feat = list(combine_df.dtypes[combine_df.dtypes == 'object'].index)
print(string_feat)
```
